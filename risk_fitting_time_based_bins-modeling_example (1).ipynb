```python
#add bins new version:learning
def add_bins_learning(s_dat, risk_seeking):
    s_dat.bin = np.full((len(s_dat.session),len(s_dat.subject_names)), -1)
    counter= -1
    for s in s_dat.subject_names:
        counter = counter+1
        subject_bins=risk_seeking.bin[risk_seeking.subject==s]
        for s_bins in subject_bins:          
            from_ms = risk_seeking.from_ms[(risk_seeking.bin==s_bins) & (risk_seeking.subject==s)]            
            from_ms.reset_index(inplace=True, drop=True)
            to_ms = risk_seeking.to_ms[(risk_seeking.bin==s_bins) & (risk_seeking.subject==s)]            
            to_ms.reset_index(inplace=True, drop=True)            
            s_dat.bin[np.where((from_ms[0]<=s_dat.feedback_time[:,counter]) & (s_dat.feedback_time[:,counter]<=to_ms[0])), counter]=s_bins
    s_dat.bin.astype(int)
    return s_dat
```

```python
#add bins new version:decision
def add_bins_decision(s_dat, risk_seeking):
    s_dat.bin = np.full((len(s_dat.session),len(s_dat.subject_names)), -1)
    counter= -1
    for s in s_dat.subject_names:
        counter = counter+1
        subject_bins=risk_seeking.bin[risk_seeking.subject==s]
        for s_bins in subject_bins:
            from_ms = risk_seeking.from_ms[(risk_seeking.bin==s_bins) & (risk_seeking.subject==s)]
            from_ms.reset_index(inplace=True, drop=True)
            to_ms = risk_seeking.to_ms[(risk_seeking.bin==s_bins) & (risk_seeking.subject==s)]
            to_ms.reset_index(inplace=True, drop=True)            
            s_dat.bin[np.where((from_ms[0]<=s_dat.stim_time[:,counter]) & (s_dat.stim_time[:,counter]<=to_ms[0])), counter]=s_bins            
    s_dat.bin.astype(int)
    return s_dat
```

```python
#add bin time new version:
def add_bin_times(s_dat, risk_seeking):
    bin_times_all=[]
    counter= -1
    for s in s_dat.subject_names:
        counter = counter+1
        subject_bins_times = []
        subject_bins=risk_seeking.bin[risk_seeking.subject==s]    
        for s_bins in subject_bins:   
            bin_times = np.array((s_dat.feedback_time[np.floor(np.median(np.where(s_dat.bin[:,counter]==s_bins))).astype(int),counter]))/ 3600. / 1000. / 24.
            subject_bins_times=np.array(np.append(subject_bins_times,bin_times))
        bin_times_all.append(subject_bins_times)
    s_dat.bin_times = bin_times_all
    return s_dat
        
```
